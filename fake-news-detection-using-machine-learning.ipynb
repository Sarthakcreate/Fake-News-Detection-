```python
import numpy as np
import pandas as pd
```

```python
true = pd.read_csv('/kaggle/input/fake-news-detection/True.csv')
fake = pd.read_csv('/kaggle/input/fake-news-detection/Fake.csv')
```

```python
fake.head()
```

```python
true.head()
```

```python
true['label']=1
fake['label']=0
```

```python
true.head()
```

```python
news = pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title','subject','date'],axis=1)
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
    text = text.lower()
    
    text = re.sub(r'https?://\S+|www\.\S+','',text)
    text = re.sub(r'<.*?>','',text)
    text = re.sub(r'[^\w\s]','',text)
    text = re.sub(r'\d','',text)
    text = re.sub(r'\n','',text)
    
    return text
    
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.transform(x_train)
xv_test =  vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()

lr.fit(xv_train,y_train)
```

```python
pred_lr = lr.predict(xv_test)
```

```python
lr.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(xv_train,y_train)
```

```python
pred_dtc = dtc.predict(xv_test)
```

```python
dtc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train,y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
```

```python
gbc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gbc))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Genuine News"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"]= new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = lr.predict(new_xv_test)
    pred_dtc = dtc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    
    return "\n\n LR Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {} ".format(
     output_label(pred_lr[0]), output_label(pred_dtc[0]), output_label(pred_rfc[0]) )
    
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
